```python
import requests
import os
import json
```

```python
documents_dir = 'documents_uploaded'

files_to_upload = []
for file_name in os.listdir(documents_dir):
    if file_name.endswith('.pdf') or file_name.endswith('.docx') or file_name.endswith('.txt'):
        file_path = os.path.join(documents_dir, file_name)
        # Read the file content and close it immediately
        with open(file_path, "rb") as f:
            file_content = f.read()
            files_to_upload.append(("files", (file_name, file_content)))

url = "http://127.0.0.1:8000/process-user-data/"

payload = {
    "user_query": "Mahatma Gandhi freedon fighter",
    "num_of_questions": 3
}




response = requests.post(url, data=payload, files=files_to_upload)



print(response.json())
```
